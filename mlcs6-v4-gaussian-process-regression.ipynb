### Parametric
Parametric methods are statistical techniques that rely on specific assumptions about the underlying distribution of the population being studied. These methods typically assume that the data follows a known Probability distribution, such as the normal distribution, and estimate the parameters of this distribution using the available data.

The basic idea behind the Parametric method is that there is a set of fixed parameters that are used to determine a probability model that is used in Machine Learning as well. Parametric methods are those methods for which we priory know that the population is normal, or if not then we can easily approximate it using a Normal Distribution which is possible by invoking the Central Limit Theorem.

**Fixed parameters*** in the context of parametric methods refer to a set of predefined values that define the shape and characteristics of a probability distribution. These parameters are assumed to be constant and unchanging throughout the dataset.

For example, in a normal distribution, the two parameters are the mean (μ) and the standard deviation (σ). Once these parameters are specified, the entire distribution is defined.

In machine learning, parametric models are often used when there's a prior assumption or belief about the underlying data distribution. The goal is to learn these fixed parameters from the data to accurately model the probability distribution.

**Regression models that follows Parametric Methods inpude**

- **Linear regression:** Predicts a continuous outcome based on a linear relationship with one or more independent variables.
- **Logistic regression:** Predicts a binary outcome (e.g., yes/no) based on a set of independent variables.
- **Naive Bayes:** Classifies data points based on Bayes’ theorem and assuming independence between features.
- **Hidden Markov Models:** Models sequential data with hidden states and observable outputs.

### Non-parametric
Non-parametric methods are statistical techniques that do not rely on specific assumptions about the underlying distribution of the population being studied. These methods are often referred to as “distribution-free” methods because they make no assumptions about the shape of the distribution.

### Imagine you're giving directions to a friend who doesn't have a map.

**Distribution-free:** They don't assume a particular shape for the underlying distribution.

**Flexible:** They can be applied to a wider range of data types and scenarios.

**Robust:** They are less sensitive to outliers and non-normality compared to parametric methods.

**Models that follows Parametric Methods inpude**

- **K-Nearest Neighbors (KNN):** Classifies data points based on the k nearest neighbors.
- **Decision Trees:** Makes classifications based on a series of yes/no questions about the features.
- **Support Vector Machines (SVM):** Creates a decision boundary that maximizes the margin between different classes.
- **Neural networks:** Can be designed with specific architectures to handle non-parametric data, such as convolutional neural networks for image data and recurrent neural networks for sequential data.


**Determining whether data is parametric or non-parametric often involves a combination of visual inspection and statistical tests.**

1. **Visual Inspection**
>- **Histograms:** Create a histogram to see the distribution of your data. If it looks roughly bell-shaped (normal), it might be parametric. If it's skewed or has multiple peaks, it's more likely to be non-parametric.
>- **Q-Q Plots:** Compare the quantiles of your data to the quantiles of a theoretical normal distribution. If the points fall close to a straight line, it suggests a parametric distribution.

2. **Statistical Tests**
>- **Shapiro-Wilk Test:** This test assesses normality. A significant p-value indicates a deviation from normality, suggesting a non-parametric distribution.
>- **Kolmogorov-Smirnov Test:** Similar to the Shapiro-Wilk test, it tests for normality.
>- **Anderson-Darling Test:*** Another test for normality, often considered more sensitive than the previous two.

3. **Skewness and Kurtosis**
>- **Skewness:** Measures the asymmetry of the distribution. A significant skew suggests a non-parametric distribution.
>- **Kurtosis:** Measures the "tailedness" of the distribution. A significant kurtosis (either positive or negative) can indicate a non-parametric distribution.

# What is Gaussian process regression?

Gaussian Process Regression (GPR) is a powerful and flexible non-parametric regression technique used in machine learning and statistics. It is particularly useful when dealing with problems involving continuous data, where the relationship between input variables and output is not explicitly known or can be complex. GPR is a Bayesian approach that can model certainty in predictions, making it a valuable tool for various applications, including optimization, time series forecasting, and more. GPR is based on the concept of a Gaussian process.

In other words: Gaussian Process Regression (GPR) is a flexible, non-parametric machine learning method used for regression tasks, meaning it predicts continuous values (like prices or temperatures). Unlike other models that assume a specific form (like linear regression assumes a straight line), GPR doesn’t assume any fixed shape for the relationship between inputs (features) and outputs (predictions). Instead, it uses probability to model many possible functions that could fit the data.

GPR doesn’t just give a single prediction; it also tells you how confident it is about that prediction.

For each new point you want to predict, it provides a mean (the predicted value) and a variance (how uncertain it is about that value).

### Simple Example:
Imagine you have data about house prices and want to predict the price of a house based on its size. GPR will consider many different possible curves that could fit your data, and it gives you the average prediction along with a confidence interval (how sure it is). If the data is sparse (ചിതറിക്കിടക്കുന്ന), the uncertainty will be higher, and if the data is dense (ഇടതിങ്ങിയ), the predictions will be more confident.

**Traditional methods:**

- **Linear regression:** Assume a straight line relationship between price and size.
- **Polynomial regression:** Assume a curved relationship.
- **Problem:** These methods assume a specific relationship, which might not always be accurate.

**GPR:**

**Multiple possibilities:** Instead of assuming one shape, GPR considers many different possible curves.

**Average prediction:** It gives you the average of all these possible predictions.

**Uncertainty:** It also tells you how confident it is in its prediction.


### Advantages of GPR:
**Flexible:** It can model complex relationships without needing to know the exact form of the data (it doesn't need to be linear).

**Uncertainty Estimates:** You get both predictions and a sense of how reliable those predictions are.

### Disadvantages of GPR:
**Computationally Slow:** It doesn’t work well with large datasets because it becomes too slow.

**Requires Good Kernel Selection:** You need to choose a good kernel function, which can be tricky.

In summary, Gaussian Process Regression is a smart and flexible way of predicting values, especially when you need to know not only the prediction but also how confident the model is about it. However, it’s best for small datasets because of its complexity.

## Importing liberarys

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from sklearn.linear_model import LinearRegression, HuberRegressor
from sklearn.isotonic import IsotonicRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

## Create the dataset

```python
A = np.linspace(start=0, stop=10, num=1_000).reshape(-1, 1)
b = np.squeeze(A * np.sin(A))
```

## Add Gaussian noise to the target values

```python
r_ng = np.random.default_rng()  # Random number generator
noise_std = 0.75
b_noisy = b + r_ng.normal(loc=0.0, scale=noise_std, size=b.shape)
```

## Split the dataset into training and testing sets

```python
A_train, A_test, b_train, b_test = train_test_split(A, b_noisy, test_size=0.2, random_state=42)
```

## Gaussian Process Regression

```python
kernel = RBF(length_scale=1.0)  # Radial Basis Function (RBF) kernel
gpr_model2 = GaussianProcessRegressor(kernel=kernel, alpha=noise_std**2)  # Set alpha as noise variance
gpr_model2.fit(A_train, b_train)
b_pred_gpr2 = gpr_model2.predict(A_test)
```

```python
# Plot Gaussian Process Regression predictions
plt.scatter(A_test, b_test, color='blue', s=15, label='Actual Test Data')
plt.plot(A_test, b_pred_gpr2, 'r.', label='GPR Predictions', linewidth=2)



# Add labels, title, and legend
plt.xlabel('A (Feature)')
plt.ylabel('b (Target)')
plt.title('Gaussian Process Regression: Predictions vs Actual')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

## Linear Regression

```python
linear_model2 = LinearRegression()
linear_model2.fit(A_train, b_train)
b_pred_linear2 = linear_model2.predict(A_test)
```

```python
# Plot Linear Regression predictions
plt.scatter(A_test, b_test, color='blue', s=5, label='Actual Test Data')
plt.plot(A_test, b_pred_linear2, color='green', label='Linear Regression Predictions')

# Add labels, title, and legend
plt.xlabel('A (Feature)')
plt.ylabel('b (Target)')
plt.title('Linear Regression: Predictions vs Actual')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

## Isotonic Regression

```python
# Isotonic regression requires a 1D array for A, hence we flatten it
iso_model2 = IsotonicRegression()
iso_model2.fit(A_train.flatten(), b_train)
b_pred_iso2 = iso_model2.predict(A_test.flatten())
```

```python
# Plot Isotonic Regression predictions
# plt.figure(figsize=(14, 8))
plt.scatter(A_test, b_test, color='blue', s=15, label='Actual Test Data')
plt.plot(A_test, b_pred_iso2, 'r.', label='Isotonic Regression Predictions', linewidth=2)

# Add labels, title, and legend
plt.xlabel('A (Feature)')
plt.ylabel('b (Target)')
plt.title('Isotonic Regression: Predictions vs Actual')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

## Huber Regression

```python
# Huber Regression (Robust to outliers)
huber_model2 = HuberRegressor()
huber_model2.fit(A_train, b_train)
b_pred_huber2 = huber_model2.predict(A_test)
```

```python
plt.scatter(A_test, b_test, color='blue', s=15, label='Actual Test Data')
plt.plot(A_test, b_pred_huber2, color='green', label='Huber Regression Predictions')

# Add labels, title, and legend
plt.xlabel('A (Feature)')
plt.ylabel('b (Target)')
plt.title('Huber Regression: Predictions vs Actual')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

## Model Evaluation

```python
# Function to evaluate the model
def evaluate_model(name, b_true, b_pred):
    mse = mean_squared_error(b_true, b_pred)
    r2 = r2_score(b_true, b_pred)
    print(f"{name}:")
    print(f"  Mean Squared Error (MSE): {mse:.4f}")
    print(f"  R² Score: {r2:.4f}\n")

# Evaluate all models
evaluate_model("Gaussian Process Regression", b_test, b_pred_gpr2)
evaluate_model("Linear Regression", b_test, b_pred_linear2)
evaluate_model("Isotonic Regression", b_test, b_pred_iso2)
evaluate_model("Huber Regression", b_test, b_pred_huber2)
```

## What is the difference  when there is noise when there is no notice in gaussian process regression 
The presence of noise in the data significantly affects the performance and behavior of Gaussian Process Regression (GPR) and other regression models.

### Model Fit
**No Noise:**

The model can perfectly fit the underlying function since the data points are exact representations of the relationship being modeled. This can lead to a model that has a very high R² score and low MSE.

**With Noise:**

The model will attempt to fit the noisy observations, which may cause it to capture not only the underlying pattern but also the noise in the data. This can lead to overfitting, where the model learns the noise as if it were a true signal, resulting in a more complex model that does not generalize well to new data.

**Generalization Performance:**

**No Noise:**

The model generally performs better on unseen data (test set) because it has learned the true relationship without being misled by random variations. The metrics like R² and MSE will be indicative of good performance.

**With Noise:**

The model may perform poorly on unseen data if it has overfitted to the noise in the training set. This can result in a lower R² score and higher MSE, as the model’s predictions may not align well with the true underlying function.

**In summary, when Gaussian Process Regression is applied to noisy data, the extra variability can make the model overfit and perform worse. On clean data, the model learns the true pattern more accurately, leading to better predictions. This shows how important data quality is when building predictive models.**
